```python
x = df[['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES',
        'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY',
        'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY',
        'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX',
        'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT',
        'TENURE']]
```

```python
df.head()
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = ss.fit_transform(x)
```

```python
# import necessary libraries
from sklearn.decomposition import PCA

from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# pca
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x)
```

```python
# kmeans algorithm
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(x)
```

```python
# agglomerative clustering algorithm
agg = AgglomerativeClustering(n_clusters=4)
agg_labels = agg.fit_predict(x)
```

```python
# plotting dandogram
plt.figure(figsize=(10,5))
linked = linkage(x, method='ward')
dendrogram(linked)
plt.title("Hierarchical Clustering Dendrogram")
plt.show()
```

```python
# plotting scatterplot
plt.figure(figsize=(6,4))
plt.scatter(X_pca[:,0], X_pca[:,1], c=kmeans_labels, cmap='viridis')
plt.title("KMeans Clusters (PCA View)")
plt.show()
```

```python
# plotting scatter
plt.figure(figsize=(6,4))
plt.scatter(X_pca[:,0], X_pca[:,1], c=agg_labels, cmap='rainbow')
plt.title("Agglomerative Clusters (PCA View)")
plt.show()
```

```python
df['KMeans_Cluster'] = kmeans_labels
df['Agglo_Cluster'] = agg_labels

print("\nCluster counts (KMeans):")
print(df['KMeans_Cluster'].value_counts())

print("\nCluster counts (Agglomerative):")
print(df['Agglo_Cluster'].value_counts())
```
